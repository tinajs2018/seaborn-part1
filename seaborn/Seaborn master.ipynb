# lineplot

```python
col1=np.linspace(0,10,1000)
col2=np.sin(col1)
df=pd.DataFrame({'c1':col1,'c2':col2})
df.head()
```

```python
# Plotting lineplot using sns.lineplot()
plt.figure(figsize=(15,5))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x=df.c1,y=df.c2,data=df)
plt.title('Line plot of the data')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.set(rc={'axes.facecolor':'pink','axes.grid':True,'xtick.labelsize':14,'ytick.labelsize':14 })
sns.lineplot(x=df.c1,y=df.c2,data=df)
sns.lineplot(x=df.c1,y=df.c2+1,data=df)
plt.title('Line plot of the data')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.set(rc={'axes.facecolor':'black','axes.grid':True,'xtick.labelsize':14,'ytick.labelsize':14 })
sns.lineplot(x=df.c1,y=df.c2,data=df,color='red')
sns.lineplot(x=df.c1,y=df.c2+1,data=df,color='white')
plt.title('Line plot of the data')
plt.show()
```
