```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
data=pd.read_csv("market.csv")
data.head()
```

```python
# fig1,axes=plt.subplots(nrows=2,ncols=20,figsize=(20,20))

# sns.swarmplot(data['Total'],data['cogs'],ax=axes[0,0],data=data,color='green')
# sns.violinplot(data['Total'],data['cogs'],ax=axes[0,0],data=data,inner=None)

# Displaying swarmplot on top of violin plot
fig1 , axes = plt.subplots(nrows=2,ncols=2 , figsize = (20,20))
sns.swarmplot(x="Quantity" , y = "Branch" , ax = axes[0,0] ,data=data, size=6 , color="#FFFFFF")
sns.violinplot(x="Quantity" , y = "Branch" , ax = axes[0,0] ,data=data , inner=None)
```

```python
#strip plot
plt.figure(figsize=(15,5))
sns.stripplot(data['cogs'])
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(data['Branch'],data['cogs'])
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(data['Branch'],data['cogs'],hue= data['Gender'])
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(data['Branch'],data['cogs'],hue= data['Gender'])
sns.boxplot(data['Branch'],data['cogs'])
```

```python
sns.boxplot(data['cogs'])
```

```python
sns.boxplot(data['cogs'],data['Branch'])
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data['cogs'],data['Branch'],hue= data['Gender'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data['cogs'],data['Branch'],hue= data['Gender'],notch=True)
plt.show()
```

```python
# Change the size of outlier markers using fliersize
plt.figure(figsize=(15,5))
sns.boxplot(data['cogs'],data['Branch'],hue= data['Gender'],notch=True,fliersize=8)
plt.show()
```
