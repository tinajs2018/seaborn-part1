```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
data=pd.read_csv("market.csv")
data.head()
```

```python
data.hist(bins=40,figsize=(20,20),color='green')
plt.show()
```

```python
plt.figure(figsize=(6,8))
x=np.linspace(0,10,100)
y=np.sin(x)
sns.kdeplot(x,y,shade=True,cmap='Reds',shade_lowest=False)
plt.show()
```

```python

plt.figure(figsize=(15,5))
sns.kdeplot(data['Tax 5%'],data['Quantity'],shade=True,cmap='Greens',shadow_lowest=False)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data['Tax 5%'],data['Quantity'],shade=False,cmap='Greens',shadow_lowest=True)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data['Tax 5%'],data['Quantity'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(data['Tax 5%'], data['Quantity'],size=10)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(data['Tax 5%'], data['Quantity'],size=10,edgecolor='black')
plt.show()
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(10,10))
sns.swarmplot(data['Tax 5%'], data['Quantity'],size=10,edgecolor='black',hue= data['Branch'])
plt.show()
```

```python

sns.catplot(x="Tax 5%" , y = "Quantity" , col= "Branch" , data=data, kind="swarm" , height=5,aspect=0.5)
```

```python
#vilon plot
sns.violinplot(data['Quantity'], data['Branch'])
```

```python
#Drawing the quartiles as horizontal lines instead of a mini-box using
sns.violinplot(data['Quantity'], data['Branch'],inner='quartile')
plt.show()
```

```python
sns.violinplot(data['Quantity'], data['Branch'],inner=None)
plt.show()
```

```python
sns.violinplot(data['Quantity'], data['Branch'],palette='muted')
plt.show()
```

```python
sns.violinplot(data['Quantity'], data['Branch'],pallette='muted',scale='count')
plt.show()
```

```python
#Show groups with different colors using "hue" and "palette"
plt.figure(figsize=(15,5))
sns.violinplot(data['Quantity'], data['Branch'],pallette='set1',hue= data['Gender'],inner='quartile')
plt.show()
```

```python
# Overlap the violinplots for different hue levels along the categorical axis using "dodge=
plt.figure(figsize=(15,5))
sns.violinplot(data['Quantity'], data['Branch'],pallette='set1',hue= data['Gender'],inner='quartile',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data['Quantity'], data['Branch'],pallette='set1',hue= data['Gender'],inner='quartile',dodge=True,split=True)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.violinplot(x=data["Quantity"] , y = data["Branch"] , palette="Set1", hue= data["Gender"],split=True)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.violinplot(x=data["Quantity"] , y = data["Branch"] , palette="Set1", hue= data["Gender"],split=True,inner="stick")
plt.show()
```

```python
# Displaying multiple violin plots using subplot function.
fig, axes=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
sns.violinplot(x=data["Quantity"] , y = data["Branch"],ax= axes[0,0] ,data=data,inner=None)
sns.violinplot(x=data["Rating"] , y = data["Branch"],ax= axes[0,1] ,data=data,inner=None)
sns.violinplot(x=data["Total"] , y = data["Branch"],ax= axes[1,0] ,data=data,inner=None)
sns.violinplot(x=data["cogs"] , y = data["Branch"],ax= axes[1,1] ,data=data,inner=None)
plt.show()
```
